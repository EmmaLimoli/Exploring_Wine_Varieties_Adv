```python
# webscrape pdfs
# import request, beautiful soup

import os
import requests 
from bs4 import BeautifulSoup
from urllib.parse import urljoin

import pprint
```

```python
# target url

url = 'http://cheeselibrary.com/making_cheese/guides'
url
```

```python
# save pdfs in folder
folder_location = r'E:\Exploring_Wine_Varieties_Adv/'
if not os.path.exists(folder_location):os.mkdir(folder_location)
```

```python
# use response and beautiful soup to create a link to pull all the pdfs from the page

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
for link in soup.select("a[href$='.pdf']"):
    filename = os.path.join(folder_location, link['href'].split('/')[-1])
    with open(filename, 'wb') as f:
        f.write(requests.get(urljoin(url, link['href'])).content)
```

```python
# import pandas and pdf reader

import PyPDF2

import pandas as pd
```

```python
# open pdf and read it

pdfFileObj = open('../Exploring_Wine_Varieties_Adv/PDFs/jg-bluepair.pdf', 'rb')

pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

print(pdfReader)
```

```python
# check pages/make sure it uploaded correctly
print(pdfReader.numPages)
```

```python
# set up pdf reader

pageObj = pdfReader.getPage(0)
pageObj
```

```python
# extract the text
print(pageObj.extractText())
```

```python
# create a variable for the extract text

page1 = pageObj.extractText()
page1
```

```python
# get rid of the slashes and add in a comma to split for the dataframe
page1 = page1.replace('\n \n',', ').replace('\n',',')
page1
```

```python
# add the names of the headers

page1 = page1.replace('Variety', 'Variety,').replace('Wines', ' Wines,').replace('Beverages', ' Beverages,').replace('Fruits/Vegetables', ' Fruits/Vegetables,').replace('Breads/Crackers',' Breads/Crackers,').replace('Condiments/Nuts', ' Condiments/Nuts,').replace('Meats/Fish', ' Meats/Fish,')
page1
```

```python
# split at the comma for the dataframe
page1 = page1.split(',')
page1
```

```python
# create the dataframe

df = pd.DataFrame([sub.split(",") for sub in page1])
df.head(50)
```

```python
# create a csv

df.to_csv('bluepair1.csv', index=False)
```
