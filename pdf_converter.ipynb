```python
# import pandas and pdf reader
import PyPDF2
import pandas as pd
```

```python
# open pdf and read it. 
# hardpair pdf
pdfFileObj = open('../Exploring_Wine_Varieties_Adv/PDFs/jg-hardpair.pdf', 'rb')

pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

print(pdfReader)
```

```python
# print number of pages
print(pdfReader.numPages)
```

```python
# read the pdf
pageObj = pdfReader.getPage(0)
pageObj
```

```python
# extract the text of the pdf
print(pageObj.extractText())
```

```python
# extract the text 
page1 = pageObj.extractText()
page1
```

```python
# replace the \n with a comma
page1 = page1.replace('\n \n',', ').replace('\n',',')
page1
```

```python

# page1 = page1.replace('Variety', 'Variety,').replace('Wines', ' Wines,').replace('Beverages', ' Beverages,').replace('Fruits/Vegetables', ' Fruits/Vegetables,').replace('Breads/Crackers',' Breads/Crackers,').replace('Condiments/Nuts', ' Condiments/Nuts,').replace('Meats/Fish', ' Meats/Fish,')
# page1
```

```python
# split to turn into a dataframe
page1 = page1.split(',')
page1
```

```python
# turn into a dataframe
df = pd.DataFrame([sub.split(",") for sub in page1])
df.head(50)
```

```python
# create a csv

df.to_csv('hardpair.csv', index=False)
```

```python
# close the pdf

pdfFileObj.close()
```
