```python
 # Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import api_key
```

```python
# Import wines_updated4.csv as DataFrame
wine_data_pd= pd.read_csv("wine_data_updated.csv")
wine_data_df = wine_data_pd.head(200)
wine_data_df
```

```python
# Add "Lat" and "Lng" columns to the Dataframe
 
wine_data_df["Lat"] = ""
wine_data_df["Lng"] = ""
wine_data_df.head(10)
```

```python
#Use Google API to find lat and lng coordinates, and add to the dataframe

target_url = "https://maps.googleapis.com/maps/api/place/textsearch/json"

for index, row in wine_data_df.iterrows():
    try:
        params = {
        "query": row["Winery"],
        "key": api_key}
        response = requests.get(target_url, params)
        
        winery_lat_lng=response.json()
        
        wine_data_df.loc[index, "Lat"] = winery_lat_lng["results"][0]["geometry"]["location"]["lat"]
        wine_data_df.loc[index, "Lng"] = winery_lat_lng["results"][0]["geometry"]["location"]["lng"]
        
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.") 
print("---------------------------------------\nData Retrieval Complete\n---------------------------------------")
```

```python
wine_data_df.to_csv("wine_lat_lng1.csv", encoding="utf-8", index=True)
```
