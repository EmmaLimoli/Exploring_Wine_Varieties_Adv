```python
# add dependencies

import pandas as pd
from sqlalchemy import create_engine
```

```python
# sqlite path

database_path = "wine_db.sqlite"
```

```python
# engine connect

engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
# read database wine

data_wine = pd.read_sql("SELECT * FROM wine", conn)
data_wine
```
